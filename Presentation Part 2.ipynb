```python
# Country_Life_by_Sex

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
data = pd.read_csv(r'C:\Users\User\Desktop\presentation 2\cleaned_Country_Life_by_Sex.csv')

# Calculate the average life expectancy for each country
avg_life_expectancy = data.groupby("Country")[["Life Expectancy of Females", "Life Expectancy for Males"]].mean()

# Sort the data by average life expectancy for males and select the top 5 countries
top_5_male = avg_life_expectancy.sort_values(by="Life Expectancy for Males", ascending=False).head(5)

# Create a bar chart for the top 5 countries
top_5_male.plot(kind='bar', figsize=(12, 6))
plt.title("Top 5 Countries with Highest Average Life Expectancy for Males")
plt.xlabel("Country")
plt.ylabel("Average Life Expectancy for Males")
plt.xticks(rotation=45)
plt.show()
```

```python
# Continents_Life_by_Sex
import pandas as pd
import altair as alt

# Load the dataset into a DataFrame
data = pd.read_csv(r'C:\Users\User\Desktop\presentation 2\cleaned_Continents_Life_by_Sex.csv')

# Create an Altair chart
chart = alt.Chart(data).mark_circle().encode(
    x='Year:T',  # Treat 'Year' as a temporal field
    y='Life Expectancy of Females:Q',
    color='Continent:N',  # Color points by 'Continent'
    tooltip=['Year:T', 'Life Expectancy of Females:Q', 'Continent:N']
).properties(
    width=600,
    height=400
)

# Display the chart
chart
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
data = pd.read_csv(r'C:\Users\User\Desktop\presentation 2\cleaned_U.S._State_Life_Expectancy_by_Sex__2020.csv')

# Calculate the average life expectancy by state and sex
avg_life_expectancy = data.groupby(["State", "Sex"])["LE"].mean().unstack()

# Select the top 5 states with the highest average life expectancy for females and males
top_5_states_female = avg_life_expectancy["Female"].nlargest(5)
top_5_states_male = avg_life_expectancy["Male"].nlargest(5)

# Combine the data for females and males
top_5_states = pd.concat([top_5_states_female, top_5_states_male])

# Create a bar chart for the top 5 states
top_5_states.plot(kind='bar', figsize=(12, 6))
plt.title("Top 5 States with Highest Average Life Expectancy in USA (Females vs. Males)")
plt.xlabel("State")
plt.ylabel("Average Life Expectancy")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
data = pd.read_csv(r'C:\Users\User\Desktop\presentation 2\cleaned_U.S._State_Life_Expectancy_by_Sex__2020.csv')

# Calculate the average life expectancy by state and sex
avg_life_expectancy = data.groupby(["State", "Sex"])["LE"].mean().unstack()

# Select the top 15 states based on average life expectancy for both males and females
top_15_states = avg_life_expectancy.mean(axis=1).nlargest(15)

# Create a bar chart for the top 15 states
top_15_states.plot(kind='bar', figsize=(12, 6))
plt.title("Top 15 States with Highest Average Life Expectancy (2020)")
plt.xlabel("State")
plt.ylabel("Average Life Expectancy")
plt.xticks(rotation=90)
plt.show()
```
